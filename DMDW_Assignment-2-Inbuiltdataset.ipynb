# Descriptive analysis on inbuilt database
https://en.wikipedia.org/wiki/Iris_flower_data_set

## Importing libraries and loading dataset

```python
import pandas as pd

# plot
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
df.rename(columns={'sepal length (cm)': 'slength', 'sepal width (cm)': 'swidth', 
                   'petal length (cm)': 'plength', 'petal width (cm)': 'pwidth'}, inplace=True)
df.head()
```

## Measuring central tendency of the data

```python
df.mean()
```

```python
df.median()
```

```python
df.mode()
```

```python
df["slength"].plot(kind="density",
              figsize=(10,10));


plt.vlines(df["slength"].mean(),     # Plot black line at mean
           ymin=0, 
           ymax=0.4,
           linewidth=1.5);

plt.vlines(df["slength"].median(),   # Plot red line at median
           ymin=0, 
           ymax=0.4, 
           linewidth=1.5,
           color="red");

plt.vlines(df["slength"].mode(),   # Plot green line at mode
           ymin=0, 
           ymax=0.4, 
           linewidth=1.5,
           color="green");
```

```python
df["swidth"].plot(kind="density",
              figsize=(10,10));


plt.vlines(df["swidth"].mean(),     # Plot black line at mean
           ymin=0, 
           ymax=1,
           linewidth=1.5);

plt.vlines(df["swidth"].median(),   # Plot red line at median
           ymin=0, 
           ymax=1, 
           linewidth=1.5,
           color="red");

plt.vlines(df["swidth"].mode(),   # Plot green line at median
           ymin=0, 
           ymax=1, 
           linewidth=1.5,
           color="green");
```

```python
df["plength"].plot(kind="density",
              figsize=(10,10));


plt.vlines(df["plength"].mean(),     # Plot black line at mean
           ymin=0, 
           ymax=0.25,
           linewidth=1.5);

plt.vlines(df["plength"].median(),   # Plot red line at median
           ymin=0, 
           ymax=0.25, 
           linewidth=1.5,
           color="red");

plt.vlines(df["plength"].mode(),   # Plot green line at median
           ymin=0, 
           ymax=0.25, 
           linewidth=1.5,
           color="green");
```

```python
df["pwidth"].plot(kind="density",
              figsize=(10,10));


plt.vlines(df["pwidth"].mean(),     # Plot black line at mean
           ymin=0, 
           ymax=0.5,
           linewidth=1.5);

plt.vlines(df["pwidth"].median(),   # Plot red line at median
           ymin=0, 
           ymax=0.5, 
           linewidth=1.5,
           color="red");

plt.vlines(df["pwidth"].mode(),   # Plot green line at median
           ymin=0, 
           ymax=0.5, 
           linewidth=1.5,
           color="green");
```

```python
features_num = ['slength', 'swidth', 'plength', 'pwidth']

# plot distributions
for f in features_num:
    fig, ax = plt.subplots(figsize=(10,7))
    
    ax.boxplot(df[f], vert=False)
    ax.grid()   
    ax.set_title('Boxplot of ' + f)
    
    plt.show()
```

## Graphical displays

```python
features_num = ['slength', 'swidth', 'plength', 'pwidth']

# plot distributions
for f in features_num:
    fig, ax = plt.subplots(figsize=(10,7))
    
    ax.hist(df[f])
    ax.grid()
    ax.set_title('Histogram of ' + f)
    
    plt.show()
```

```python
plt.scatter(df['slength'], df['plength'])
plt.title('sepal length vs petal length')
plt.xlabel('slength')
plt.ylabel('plength')
plt.grid()
plt.show()
```

```python
plt.scatter(df['swidth'], df['pwidth'])
plt.title('sepal width vs petal width')
plt.xlabel('swidth')
plt.ylabel('pwidth')
plt.grid()
plt.show()
```
